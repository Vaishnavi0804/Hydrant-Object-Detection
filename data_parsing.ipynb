```python
import xml.etree.ElementTree as ET
import pandas as pd

# Parse the XML file
tree = ET.parse('#########file path##########')
root = tree.getroot()

# Create an empty list to store the data
data = []

# Extract the required information from the XML file
filename = root.find('filename').text

for obj in root.findall('object'):
    name = obj.find('name').text
    xmin = float(obj.find('bndbox/xmin').text)
    ymin = float(obj.find('bndbox/ymin').text)
    xmax = float(obj.find('bndbox/xmax').text)
    ymax = float(obj.find('bndbox/ymax').text)
    
    # Extract attributes
    attributes = obj.find('attributes')
    if attributes is not None:
        for attr in attributes.findall('attribute'):
            attr_name = attr.find('name').text
            attr_value = attr.find('value').text
            data.append([filename, name, xmin, ymin, xmax, ymax, attr_name, attr_value])
    else:
        data.append([filename, name, xmin, ymin, xmax, ymax, '', ''])

# Create a Pandas dataframe
df = pd.DataFrame(data, columns=['filename', 'object', 'xmin', 'ymin', 'xmax', 'ymax', 'attribute', 'value'])
```

```python
import os
import xml.etree.ElementTree as ET
import pandas as pd

# Create an empty list to store the data
data = []

# Loop through all the XML files in the folder
folder = 'folder path'  # Replace with the name of your folder
for filename in os.listdir(folder):
    if filename.endswith('.xml') and filename.startswith('test'):
        # Parse the XML file
        tree = ET.parse(os.path.join(folder, filename))
        root = tree.getroot()

        # Extract the required information from the XML file
        image_filename = filename[:-4] + '.jpg'

        for obj in root.findall('object'):
            name = obj.find('name').text
            xmin = float(obj.find('bndbox/xmin').text)
            ymin = float(obj.find('bndbox/ymin').text)
            xmax = float(obj.find('bndbox/xmax').text)
            ymax = float(obj.find('bndbox/ymax').text)

            # Extract attributes
            attributes = obj.find('attributes')
            if attributes is not None:
                for attr in attributes.findall('attribute'):
                    attr_name = attr.find('name').text
                    attr_value = attr.find('value').text
                    data.append([image_filename, name, xmin, ymin, xmax, ymax, attr_name, attr_value])
            else:
                data.append([image_filename, name, xmin, ymin, xmax, ymax, '', ''])

# Create a Pandas dataframe
df = pd.DataFrame(data, columns=['filename', 'object', 'xmin', 'ymin', 'xmax', 'ymax', 'attribute', 'value'])
```
